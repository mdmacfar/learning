```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://coinmarketcap.com/trending-cryptocurrencies/").text

#create beautifulsoup object
soup = BeautifulSoup(page, "html.parser")

#pull all instances of <a> tag
cryptos = soup.find_all('a')

#clears data of all tags
for crypto in cryptos:
    names = crypto.contents[0]
    fullLink = crypto.get('href')
    print(names)
    print(fullLink)
```
